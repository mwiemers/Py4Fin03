### Examples from the introduction

#### Plotting multiple lines

````python
import pandas_datareader.data as web
import matplotlib.pyplot as plt

azn = web.DataReader('AZN', 'yahoo', start='2020')
pfe = web.DataReader('PFE', 'yahoo', start='2020')

plt.plot(azn['Adj Close'], label='Astra-Zeneca')
plt.plot(pfe['Adj Close'], label='Pfizer')

plt.legend()
plt.show()
````

#### Plotting area instead of line

````python
plt.fill_between(azn.index, azn['Adj Close'])
plt.fill_between(pfe.index, pfe['Adj Close'])

plt.show()
````

#### Adding a legend

````python
plt.fill_between(azn.index, azn['Adj Close'], label='Astra-Zaneca')
plt.fill_between(pfe.index, pfe['Adj Close'], label='Pfizer')

plt.legend(loc='upper left')
plt.show()
````

### Bonus task - Further customizations

You can change default parameters of matplotlib using the plt.rcParams dictionary.

For instance, you can change the linewidth with the key 'lines.linewidth'.

```python
plt.rcParams['lines.linewidth'] = 1.5
```

You can look up the names of all runtime configuration parameters with:

````python
plt.rcParams.keys()
````

To reset all parameters to default, you can run the following line

```python
plt.style.use('default')
```
