### Examples from the introduction

#### Simple line plot

```python
import matplotlib.pyplot as plt
import pandas_datareader.data as web

btc = web.DataReader('BTC-USD', 'yahoo', '1-1-2000')

plt.plot(btc.index, btc['Close'])
plt.show()
```

#### Applying customizations

```python
plt.plot(btc['Adj Close'])
plt.title('Bitcoin') 
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.yticks([0, 10000, 20000, 30000, 40000, 50000, 60000],
           ['$0', '$10000', '$20000', '$30000', '$40000', '$50000', '$60000'])
plt.show()
```

### Changing the color, linewidth and alpha value

You can change the colour of the line using the **color parameter**. Here, we pass the value `'tab:orange'`, which is a specific orange. We can change the width of the line with the **linewidth** parameter. The **alpha** parameter allows you to change the transparency of the line. Setting this to a level below 1 usually makes the plot look a bit nicer.

```python
plt.plot(btc['Adj Close'], color='tab:orange', linewidth=1.5, alpha=.7)
plt.title('Bitcoin') 
plt.xlabel('Date')
plt.ylabel('Price in USD')
plt.yticks([0,10000,20000,30000,40000,50000,60000],
           ['$0','$10000','$20000','$30000','$40000','$50000','$60000'])
plt.show()
```
